```python
#import python libraries and csv file
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\SUMERINA\Downloads\COGNIFYZ_Sumerina\Dataset .csv", encoding= 'unicode_escape')
```

Task- 1

Analyze the distribution of aggregate
ratings and determine the most common
rating range.

```python
agg_ratings_distribution=df['Aggregate rating'].value_counts()
print(agg_ratings_distribution)
```

```python
a=agg_ratings_distribution.idxmax()
print("Most common rating range is : ",a)
```

```python
plt.hist(df['Aggregate rating'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel("Aggregate rating")
plt.ylabel("Count")
plt.title("Distribution of aggregate rating")
plt.show()
```

Calculate the average number of votes
received by restaurants.

```python
avg_vote=round(df['Votes'].mean(),3)
print("The average number of votes received by restaurants : ", avg_vote)
```
